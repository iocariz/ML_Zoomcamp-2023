```python
import pandas as pd
```

```python
data = pd.read_csv('happyscore_income.csv')
```

```python
data
```

```python
data.sort_values(by='avg_income',  inplace=True)
```

```python
richest = data.tail(10)
```

```python
#plot avg_income vs happyScore and label points with country 

import matplotlib.pyplot as plt
import numpy as np

plt.scatter(richest['avg_income'], richest['happyScore'])

for k, row in richest.iterrows():
    plt.annotate(row['country'], xy=(row['avg_income'], row['happyScore']))

plt.xlabel('Average Income')
plt.ylabel('Happiness Score')
plt.title('TOP 10 Richest countries: Happiness vs Income')
plt.show()



```
